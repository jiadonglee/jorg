```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import sys
sys.path.append('/Users/jdli/Project/Korg.jl/Jorg/src/jorg/')

```

```python
eos_compare_df = pd.read_csv("/Users/jdli/Project/Korg.jl/temp_analysis/jorg_korg_grid_validation_standard_20250707_225853.csv")

eos_compare_df.shape, eos_compare_df.columns
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8), gridspec_kw={'height_ratios': [3, 1]}, sharex=True)

# Top panel: Electron density solutions
axs[0].plot(eos_compare_df['Teff'], eos_compare_df['jorg_ne_solution'], 'o', label='NE Solution: Jorg', markersize=3)
axs[0].plot(eos_compare_df['Teff'], eos_compare_df['korg_ne_solution'], '^', label='NE Solution: Korg', markersize=3)
axs[0].set(yscale='log', ylabel='Electron Density (cm$^{-3}$)', title='Electron Density Comparison: Jorg vs Korg')
axs[0].legend()

# Bottom panel: Relative errors
axs[1].plot(eos_compare_df['Teff'], eos_compare_df['diff_ne_solution_percent'], 'o', color='green', markersize=3)
axs[1].axhline(y=0, color='r', linestyle='-', alpha=0.3)
axs[1].set(xlabel='Teff (K)', ylabel='Difference (%)')

plt.tight_layout()
```
